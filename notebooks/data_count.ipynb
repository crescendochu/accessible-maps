```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import MultiPolygon, Polygon
```

```python
def count_points(geometry):
    count = 0
    if isinstance(geometry, MultiPolygon):
        # Correctly iterate over each Polygon in a MultiPolygon
        for polygon in geometry.geoms:
            # Exterior ring points
            count += len(polygon.exterior.coords)
            # Interior rings (holes) points
            for interior in polygon.interiors:
                count += len(interior.coords)
    elif isinstance(geometry, Polygon):
        # Handle case where geometry is a single Polygon
        count += len(geometry.exterior.coords)
        for interior in geometry.interiors:
            count += len(interior.coords)
    # Add other geometry types handling if necessary
    return count
```

```python
#read in the data
state = gpd.read_file('../data/processed/ppl_density/00_state/state_ppl_density.shp')
```

```python
state
```

```python
state['points_count'] = state['geometry'].apply(count_points)
```

```python
state['points_count'].sum()
```
